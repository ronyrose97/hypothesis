# Testing wheather sales increase after digital marketing

```python
import scipy.stats as stats
```

H0:Mean sales after stepping into digital marketting>Mean Sales before stepping into digital marketing

H1:Mean Sales after stepping into digital marketing<=Mean Sales before stepping into digital marketing

```python
ttest,pval = stats.ttest_ind(data['Sales_After_digital_add(in $)'],
                             data['Sales_before_digital_add(in $)'],alternative="greater")
print("p-value",f"{pval:.27f}")
if pval <0.05:
  print("We reject null hypothesis.")
else:
  print("We accept null hypothesis.")
               
```

# Dependency of manager and region

```python
manager_number={'Manager - A':1,'Manager - B':2,'Manager - C':3}
region_number={'Region - A':1,'Region - B':2,'Region - C':3}
```

```python
Manager_Num = [manager_number[item] for item in data.Manager]
Region_Num = [region_number[item] for item in data.Region]
```

```python
Manager_Num
```

```python
Region_Num
```

```python
data['Manager_Num']=Manager_Num
data['Region_Num']=Region_Num
data
```

```python
from scipy.stats import chi2_contingency
table = data.drop(['Month','Manager','Region','Sales_before_digital_add(in $)','Sales_After_digital_add(in $)'],axis=1)
stat, p, dof, expected = chi2_contingency(table)
print('stat=%.3f, p=%.3f' % (stat, p))
if p < 0.05:
    print('The variables Region and Manager are probably dependent')
else:
    print('The variables Region and Manager are probably independent')
```
